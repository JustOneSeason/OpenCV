```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/Colab/Images/girl.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
plt.imshow(gimg, cmap = 'gray')
```

```python
gPlus = np.zeros((gimg.shape[0], gimg.shape[1]), dtype = np.ubyte)
```

```python
val = 50

for h in range(gimg.shape[0]):
    for w in range(gimg.shape[1]):
      plusVal = gimg[h,w] + val
      if(plusVal > 255):
        plusVal = 255
      gPlus[h,w] = plusVal

plt.imshow(gPlus, cmap = 'gray')
```

```python
img = cv2.imread('/content/gdrive/My Drive/Colab/Images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg2 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
cv2_imshow(gimg2)
```

```python
for h in range(gimg1.shape[0]):
    for w in range(gimg1.shape[1]):
      plusVal = np.int32(gimg1[h,w]) + np.int32(gimg2[h,w])
      if(plusVal > 255):
        plusVal = 255
      gPlus[h,w] = plusVal

cv2_imshow(gPlus)
```

```python
weight = 0.7


for h in range(gimg1.shape[0]):
    for w in range(gimg1.shape[1]):
      plusVal = np.int32(gimg1[h,w])*weight + np.int32(gimg2[h,w])*(1-weight)
      if(plusVal > 255):
        plusVal = 255
      gPlus[h,w] = plusVal

cv2_imshow(gPlus)
```
