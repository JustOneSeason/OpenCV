```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/Colab/Images/lena256.jpg')
#img = cv2.imread('/content/gdrive/My Drive/Colab/Images/fce(salt_pepper noise).bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
plt.imshow(gimg, cmap = 'gray')
```

```python
outimg = cv2.GaussianBlur(gimg,(7,7),1)

cv2_imshow(outimg)
```

```python
outimg = cv2.GaussianBlur(gimg,(7,7),3)

cv2_imshow(outimg)
```

```python
outimg = cv2.GaussianBlur(gimg,(15,15),1)

cv2_imshow(outimg)
```

```python
outimg = cv2.GaussianBlur(gimg,(15,15),3)

cv2_imshow(outimg)
```

```python
mean_img = cv2.blur(gimg,(5,5))
cv2_imshow(mean_img)
```

```python
edgeimg = cv2.addWeighted(gimg,1.0,mean_img,-1.0,0)
cv2_imshow(edgeimg)
```

```python
out_img = cv2.addWeighted(gimg, 1.0, edgeimg, 3.0,0)
cv2_imshow(out_img)
```
